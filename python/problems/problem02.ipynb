# 모든코인 상승장? 하락장?
> 최고가와 최저가의 차이를 변동폭으로 정의할 때 (시가 + 변동폭)이 최고가 보다 높을 경우 "상승장", 그렇지 않은 경우 "하락장" 문자열을 출력하라.


|Key Name|Description|
|------|---|
|opeing_price|최근 24시간 내 시작 거래금액|
|closing_price|최근 24시간 내 마지막 거래금액|
|min_price|최근 24시간 내 최저 거래금액|
|max_price|최근 24시간 내 최고 거래금액|

```python
import requests, pprint
url = "https://api.bithumb.com/public/ticker/all"
data = requests.get(url).json()['data']
pprint.pprint(data)
```

```python
for key in data.keys():
    if key == 'date':
        continue
    elif float(data[f'{key}']['max_price']) - float(data[f'{key}']['min_price']) + float(data[f'{key}']['opening_price']) > float(data[f'{key}']['max_price']):
        print(f'{key} : 상승장')
    else:
        print(f'{key} : 하락장')
```

```python
for coin, coin_data in data.items():
#     if type(coin_data) != type(dict()):
#         continue
    if type(coin_data) == type(dict()):
        if float(coin_data['opening_price']) - float(coin_data['min_price']) > 0:
            print('상승장')
        else:
            print('하락장')
```

# 평균점수
> 다음 딕셔너리에서 평균 점수를 출력하라


```python
student = {'python':80, 'algorithm':78, 'django':95, 'flask':80}
p = 0
result = 0
for score in student.values():
    result += score
    p += 1
print(f'평균 점수 : {result/p}')
```

# 혈액형
> 학생들의 혈액형(A,B,O,AB)에 대한 데이터가 리스트에 들어있다. 각 혈액형이 몇명인지 딕셔너리를 만들어 출력하라

```python
blood = ['A','A','B','O','A','B','A','AB','AB','O','A','O','AB','O']
a = 0
b = 0
o = 0
ab = 0
for blood_type in blood:
    if blood_type == 'A':
        a += 1
    elif blood_type == 'B':
        b += 1
    elif blood_type == 'O':
        o += 1
    elif blood_type == 'AB':
        ab += 1
dic = { 'A형': f'{a}명', 'B형': f'{b}명', 'O형': f'{o}명', 'AB형':f'{ab}명'}
print(dic)
```

```python
blood = ['A','A','B','O','A','B','A','AB','AB','O','A','O','AB','O']

dic = {}

for i in blood:
    if i not in dic:
        dic[i] = 1
    else:
        dic[i] += 1
print(dic)
#     if dic[i] in i:
#         print(dic)
```

```python
blood_dict = {}

# (반복)하나씩 꺼내서 확인한다.
for blood_type in blood:
    # 지금 나온적이 없으면, 키와 value 1로 만든다.
    if blood_type not in blood_dict.keys():
        blood_dict[blood_type] = 1
    else:
        blood_dict[blood_type] += 1
print(blood_dict)
```

# UBD
> movies는 영화제목이 key로 누적관객수가 value인 딕셔너리이다. 
>
> 자전차왕 엄복동의 누적관객수는 172212명이고 172212명을 1UBD라고 할때 80UBD를 넘지 못하는 영화를 출력하라.

```python
movies = {
    "7번방의선물":12811206,
    "괴물":13019740,
    "국제시장":14257115,
    "극한직업":16261018,
    "도둑들":12983330,
    "명량":17613682,
    "베테랑":13414009,
    "신과함께-죄와벌":14410754,
    "아바타":13624328,
    "어벤져스:엔드게임":13901423,
}

UBD = 172212
for key, value in movies.items():
    if value < 80 * UBD:
        print(key)
```
